```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
songs = pd.read_csv("songs_2020.csv")
len(songs)
```

```python
songs.drop_duplicates(subset="spotify_id", keep='first', inplace=True)
len(songs)
```

```python
songs.drop(["Unnamed: 0"], inplace=True, axis=1)
songs.head()
```

```python
mask = songs["name"].str.contains("hello", case=False, regex=False)
songs[mask]
```

```python
songs.hist(column = "popularity", bins=10, figsize=(16, 10))
```

```python
songs.drop(songs[(songs["popularity"] <40)].index, axis=0, inplace=True)
```

```python
songs.to_csv("processed/2020.csv")
```
